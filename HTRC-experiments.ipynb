```python
import pandas as pd
```

```python
# I got these CSV files from https://analytics.hathitrust.org/datasets
poetry = pd.read_csv('poetry_yearly_summary.csv')
fiction = pd.read_csv('fiction_yearly_summary.csv')
```

```python
# First, let's look at words irrespective of year. 
```

```python
poetrySums = poetry.groupby('word').sum()
fictionSums = fiction.groupby('word').sum()
```

```python
(poetrySums['termfreq']-fictionSums['termfreq']).fillna(0).sort_values()
```

```python
def century(year): 
    return str((year // 100) + 1) + 'th'
```

```python
poetry['century'] = poetry['year'].apply(century)
fiction['century'] = fiction['year'].apply(century)
```

```python
def groupByCenturyAndWord(df): 
    sums = {}
    for century, data in df.groupby('century'):
        sums[century] = data.groupby('word').sum()
    return sums
    
```

```python
poetryCenturiesWords = groupByCenturyAndWord(poetry)
fictionCenturiesWords = groupByCenturyAndWord(fiction)
```

```python
for century in ['18th', '19th', '20th']: 
    print('------%s------' % century, (poetryCenturiesWords[century]-fictionCenturiesWords[century]).fillna(0).sort_values(by='termfreq'))
```
